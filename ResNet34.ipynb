```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Conv2D , MaxPooling2D , BatchNormalization , GlobalAveragePooling2D
from tensorflow.keras.models import save_model,Model
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
import numpy as np
```

# Load Dataset

```python
dataset , info = tfds.load("stanford_online_products", split = "train"  ,with_info = True)
```

```python
ds_test = tfds.load("stanford_online_products", split = "test")
```

```python
info
```

```python
fig = tfds.show_examples(ds = dataset.take(12), ds_info=info)
```

# Preprocessing

```python
batch_size = 128
def preprocess_data(example):
    image = example['image']
    image = tf.image.resize(image, [224, 224])
    image = tf.cast(image, tf.float32) / 255.0
    label = example['super_class_id']
    return image, label
```

```python
dataset = dataset.map(preprocess_data)
ds_test = ds_test.map(preprocess_data)

dataset = dataset.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
ds_test = ds_test.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
```

# Build ResNet34 model

```python
class ResNet34(Model):
    def __init__(self , n_class):
        super(ResNet34, self).__init__()
        
        self.conv = Conv2D(64, 7, strides = 2, padding = 'same')
        self.b = BatchNormalization()
        self.max_pool = MaxPooling2D(3, strides = 2)
        
#         ResBlock1
        self.conv1_1 = Conv2D(64, 3, strides = 1 , padding = 'same')
        self.b1_1 = BatchNormalization()
        self.conv1_2 = Conv2D(64, 3, strides = 1, padding = 'same')
        self.b1_2 = BatchNormalization()
        
#         ResBlock2
        self.conv2_1 = Conv2D(64, 3, strides = 1 , padding = 'same')
        self.b2_1 = BatchNormalization()
        self.conv2_2 = Conv2D(64, 3, strides = 1, padding = 'same')
        self.b2_2 = BatchNormalization() 
        
#         ResBlock3
        self.conv3_1 = Conv2D(64, 3, strides = 1 , padding = 'same')
        self.b3_1 = BatchNormalization()
        self.conv3_2 = Conv2D(64, 3, strides = 1, padding = 'same')
        self.b3_2 = BatchNormalization() 
        
#         ResBlock4
        self.conv4_1 = Conv2D(128, 3, strides = 2 , padding = 'same')
        self.b4_1 = BatchNormalization()
        self.conv4_2 = Conv2D(128, 3, strides = 1, padding = 'same')
        self.b4_2 = BatchNormalization() 
        
        self.conv4_res = Conv2D(128, 1, strides = 2)
        self.b4_res = BatchNormalization()
        
#         ResBlock5
        self.conv5_1 = Conv2D(128, 3, strides = 1 , padding = 'same')
        self.b5_1 = BatchNormalization()
        self.conv5_2 = Conv2D(128, 3, strides = 1, padding = 'same')
        self.b5_2 = BatchNormalization()        

#         ResBlock6
        self.conv6_1 = Conv2D(128, 3, strides = 1 , padding = 'same')
        self.b6_1 = BatchNormalization()
        self.conv6_2 = Conv2D(128, 3, strides = 1, padding = 'same')
        self.b6_2 = BatchNormalization()  

#         ResBlock7
        self.conv7_1 = Conv2D(128, 3, strides = 1 , padding = 'same')
        self.b7_1 = BatchNormalization()
        self.conv7_2 = Conv2D(128, 3, strides = 1, padding = 'same')
        self.b7_2 = BatchNormalization() 
        
#         ResBlock8
        self.conv8_1 = Conv2D(256, 3, strides = 2 , padding = 'same')
        self.b8_1 = BatchNormalization()
        self.conv8_2 = Conv2D(256, 3, strides = 1, padding = 'same')
        self.b8_2 = BatchNormalization() 
        
        self.conv8_res = Conv2D(256, 1, strides = 2)
        self.b8_res = BatchNormalization()
        
#         ResBlock9
        self.conv9_1 = Conv2D(256, 3, strides = 1 , padding = 'same')
        self.b9_1 = BatchNormalization()
        self.conv9_2 = Conv2D(256, 3, strides = 1, padding = 'same')
        self.b9_2 = BatchNormalization() 
        
#         ResBLock10
        self.conv10_1 = Conv2D(256, 3, strides = 1 , padding = 'same')
        self.b10_1 = BatchNormalization()
        self.conv10_2 = Conv2D(256, 3, strides = 1, padding = 'same')
        self.b10_2 = BatchNormalization() 
        
#         ResBlock11
        self.conv11_1 = Conv2D(256, 3, strides = 1 , padding = 'same')
        self.b11_1 = BatchNormalization()
        self.conv11_2 = Conv2D(256, 3, strides = 1, padding = 'same')
        self.b11_2 = BatchNormalization() 
        
#         ResBLock12
        self.conv12_1 = Conv2D(256, 3, strides = 1 , padding = 'same')
        self.b12_1 = BatchNormalization()
        self.conv12_2 = Conv2D(256, 3, strides = 1, padding = 'same')
        self.b12_2 = BatchNormalization()  
        
#         ResBlock13
        self.conv13_1 = Conv2D(256, 3, strides = 1 , padding = 'same')
        self.b13_1 = BatchNormalization()
        self.conv13_2 = Conv2D(256, 3, strides = 1, padding = 'same')
        self.b13_2 = BatchNormalization()
        
#         ResBLock14
        self.conv14_1 = Conv2D(512, 3, strides = 2 , padding = 'same')
        self.b14_1 = BatchNormalization()
        self.conv14_2 = Conv2D(512, 3, strides = 1, padding = 'same')
        self.b14_2 = BatchNormalization() 
        
        self.conv14_res = Conv2D(512, 1, strides = 2)
        self.b14_res = BatchNormalization()
        
#         ResBLock15
        self.conv15_1 = Conv2D(512, 3, strides = 1 , padding = 'same')
        self.b15_1 = BatchNormalization()
        self.conv15_2 = Conv2D(512, 3, strides = 1, padding = 'same')
        self.b15_2 = BatchNormalization()  
        
#         ResBLock16
        self.conv16_1 = Conv2D(512, 3, strides = 1 , padding = 'same')
        self.b16_1 = BatchNormalization()
        self.conv16_2 = Conv2D(512, 3, strides = 1, padding = 'same')
        self.b16_2 = BatchNormalization() 
        
        
        self.activation = tf.keras.layers.Activation(tf.nn.relu)
        self.avg_pool = GlobalAveragePooling2D()

        self.fc = Dense(n_class , activation='softmax')

    def call(self, x):
        x = self.conv(x)
        x = self.b(x)
        x = self.activation(x)
        
        x = self.max_pool(x)
        
#         ResBlock1
        x1 = self.conv1_1(x)
        x1 = self.b1_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv1_2(x1)
        x1 = self.b1_2(x1)
        x1 = self.activation(x1)
        
        x = tf.keras.layers.Add()([x1, x])
        
#         ResBlock2
        x1 = self.conv2_1(x)
        x1 = self.b2_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv2_2(x1)
        x1 = self.b2_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x])
        
#         ResBlock3
        x1 = self.conv3_1(x)
        x1 = self.b3_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv3_2(x1)
        x1 = self.b3_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x])       

#         ResBlock4
        x1 = self.conv4_1(x)
        x1 = self.b4_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv4_2(x1)
        x1 = self.b4_2(x1)
        x1 = self.activation(x1)
        
        x2 = self.conv4_res(x)
        x2 = self.b4_res(x2)

        x = tf.keras.layers.Add()([x1, x2]) 
        
#         ResBlock5
        x1 = self.conv5_1(x)
        x1 = self.b5_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv5_2(x1)
        x1 = self.b5_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock6
        x1 = self.conv6_1(x)
        x1 = self.b6_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv6_2(x1)
        x1 = self.b6_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock7
        x1 = self.conv7_1(x)
        x1 = self.b7_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv7_2(x1)
        x1 = self.b7_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock8
        x1 = self.conv8_1(x)
        x1 = self.b8_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv8_2(x1)
        x1 = self.b8_2(x1)
        x1 = self.activation(x1)
        
        x2 = self.conv8_res(x)
        x2 = self.b8_res(x2)

        x = tf.keras.layers.Add()([x1, x2]) 
        
#         ResBlock9
        x1 = self.conv9_1(x)
        x1 = self.b9_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv9_2(x1)
        x1 = self.b9_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock10
        x1 = self.conv10_1(x)
        x1 = self.b10_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv10_2(x1)
        x1 = self.b10_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 

#         ResBlock11
        x1 = self.conv11_1(x)
        x1 = self.b11_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv11_2(x1)
        x1 = self.b11_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock12
        x1 = self.conv12_1(x)
        x1 = self.b12_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv12_2(x1)
        x1 = self.b12_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock13
        x1 = self.conv13_1(x)
        x1 = self.b13_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv13_2(x1)
        x1 = self.b13_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock14
        x1 = self.conv14_1(x)
        x1 = self.b14_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv14_2(x1)
        x1 = self.b14_2(x1)
        x1 = self.activation(x1)
        
        x2 = self.conv14_res(x)
        x2 = self.b14_res(x2)

        x = tf.keras.layers.Add()([x1, x2]) 
        
#         ResBlock15
        x1 = self.conv15_1(x)
        x1 = self.b15_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv15_2(x1)
        x1 = self.b15_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x]) 
        
#         ResBlock16
        x1 = self.conv16_1(x)
        x1 = self.b16_1(x1)
        x1 = self.activation(x1)
        
        x1 = self.conv16_2(x1)
        x1 = self.b16_2(x1)
        x1 = self.activation(x1)

        x = tf.keras.layers.Add()([x1, x])
        
        x = self.avg_pool(x)
        x = self.fc(x)
        return x
    
    def model(self):
        x = tf.keras.layers.Input(shape=(224, 224, 3))
        print(self.call(x))
        return Model(inputs=[x], outputs=self.call(x))
```

# Summary

```python
model = ResNet34(12)
model.model().summary()
```

```python
loss_object = tf.keras.losses.SparseCategoricalCrossentropy()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
train_loss = tf.keras.metrics.Mean(name='train_loss')
train_accuracy = tf.keras.metrics.SparseCategoricalAccuracy(name='train_accuracy')

test_loss = tf.keras.metrics.Mean(name='test_loss')
test_accuracy = tf.keras.metrics.SparseCategoricalAccuracy(name='test_accuracy')
```

```python
@tf.function
def train_step(x_train, y_train):
    with tf.GradientTape() as tape:
        predictions = model(x_train, training=True)
        loss = loss_object(y_train, predictions)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    train_loss(loss)
    train_accuracy(y_train, predictions)
```

```python
@tf.function
def test_step(images, labels):
    predictions = model(images, training=False)
    t_loss = loss_object(labels, predictions)

    test_loss(t_loss)
    test_accuracy(labels, predictions)
```

# Train

```python
EPOCHS = 5
for epoch in range(EPOCHS):
    train_loss.reset_states()
    train_accuracy.reset_states()
    test_loss.reset_states()
    test_accuracy.reset_states()
    
    for images  ,labels in dataset:
        
        train_step(images , labels)
        
    for images  ,labels in ds_test:
        
        test_step(images , labels)

    print(
    f'Epoch {epoch + 1}, '
    f'Loss: {train_loss.result()}, '
    f'Accuracy: {train_accuracy.result()}, '
    f'ValLoss: {test_loss.result()}, '
    f'ValAccuracy: {test_accuracy.result()}, '
  )
```
